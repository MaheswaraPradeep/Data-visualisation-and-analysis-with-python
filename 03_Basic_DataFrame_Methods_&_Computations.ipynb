# Basic DataFrame Analysis

* `sum`
* `min`
* `max`
* `median`
* `mean`
* `count`
* `describe`

```python
import pandas as pd
```

```python
houses=pd.read_csv("/content/drive/MyDrive/udemy pandas csv data /kc_house_data.csv")
houses
```

```python
houses.min() #gives min of all cols
```

```python
houses.max() #give max of all cols
```

```python
type(houses.max()) #the type is a series
```

```python
houses.sum() #gives sum of all values in each col
```

```python
houses.sum(numeric_only=True) #does sum of only numeric cols not objects or strings
```

```python
titanic=pd.read_csv("/content/drive/MyDrive/udemy pandas csv data /titanic.csv")
```

```python
titanic.sum()
```

```python
titanic.sum(numeric_only=True)
```

count method

```python
titanic.count() #gives total no of non null vals in each col
```

Mean,median,mode

```python
titanic.mean()
```

```python
houses.median()
```

```python
titanic.median()
```

```python
titanic.mode() #gives most frequent value for each col
```

Describe method

```python
titanic.describe() #gives basic statistic description of each col
```
